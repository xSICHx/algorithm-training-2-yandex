```python
class binary_tree:

    class mem_struct:
        def __init__(self, n):
            self.array = [[None, i+1, None] for i in range(n)]
            self.first_empty = 0

        def new_node(self):
            self.first_empty = self.array[self.first_empty][1]
        
        def del_node(self, index):
            self.array[index][1] = self.first_empty
            self.first_empty = index
        
        def get_key(self, i):
            return self.array[i][0]
        
        def get_left(self, i):
            return self.array[i][1]
        
        def get_right(self, i):
            return self.array[i][2]

        def set_key(self, i, val):
            self.array[i][0] = val
        
        def set_left(self, i, val):
            self.array[i][1] = val
        
        def set_right(self, i, val):
            self.array[i][2] = val
        
    
    def __init__(self, n):
        self.mem = self.mem_struct(n)
        self.root = 0

    
    def find(self, x, root = None):
        if root == None:
            root = self.root
        key = self.mem.get_key(root)
        if x == key:
            return root
        elif x < key:
            left = self.mem.get_left(root)
            if left == None:
                return None
            else:
                return self.find(x, left)
        else:
            right = self.mem.get_right(root)
            if right == None:
                return None
            else:
                return self.find(x, right)
    

    def create_and_fill_node(self, key):
        self.mem.new_node()
        ind = self.mem.first_empty
        self.mem.set_key(ind, key)
        self.mem.set_left(ind, None)
        self.mem.set_right(ind, None)
        return ind
    

    def add(self, x, root = None):
        if root == None:
            root = self.root

        key = self.mem.get_key(root)
        if key == None:
            self.create_and_fill_node(x)
            return
            
        if x < key:
            left = self.mem.get_left(root)
            if left == None:
                self.mem.set_left(root, self.create_and_fill_node(x))
            else:
                self.add(left, x)
        elif x > key:
            right = self.mem.get_right(root)
            if right == None:
                self.mem.set_right(root, self.create_and_fill_node(x))
            else:
                self.add(right, x)
    
    
    def print(self, root = None):
        if root == None:
            root = self.root

        if self.mem.get_left(root) != None:
            self.print(self.mem.get_left(root))
        if self.mem.get_key(root) != None:
            print(self.mem.get_key(root), end = " ")
        if self.mem.get_right(root) != None:
            self.print(self.mem.get_right(root))
    

        

tree = binary_tree(10)
tree.add(10)
tree.add(2)
tree.add(11)
tree.add(15)
tree.add(13)
tree.add(4)
tree.add(1)
tree.print()


```
