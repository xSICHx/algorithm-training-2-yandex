```python
beg_segment_const = 1
end_segment_const = -1


def update_begin_segment_with_max_end(max_var, curr):
        if max_var == None:
            if curr[1] > 0:
                return curr
            else:
                return None
        else:
            if curr[1] > max_var[1]:
                return curr
            else:
                return max_var


def update_end_segment_with_min_beg(min_var, curr, m):
        if min_var == None:
            if curr[0] < m:
                return curr
            else:
                return None
        else:
            if curr[0] < min_var[0]:
                return curr
            else:
                return min_var


def find_segments():
    # read
    file = open("input.txt", "r")
    
    m = int(file.readline())
    points = []
    segments = []
    i = 0
    while True:
        segment = tuple(map(int, file.readline().strip().split()))
        if segment == (0, 0):
            break
        points.append( (segment[0], beg_segment_const, i) )
        points.append( (segment[1], end_segment_const, i) )
        segments.append(segment)
        i += 1
    points.sort()
    
    file.close()
    
    # actions
    counter = 0
    begin_segment_with_max_end = None
    end_segment_with_min_beg = None
    last_beg = 0
    res = set()
    for point in points:
        # condition when continue (not in [0, m])
        index_seg = point[2]
        seg = segments[index_seg]
        if seg[0] <= 0 and seg[1] <= 0:
            continue
        if seg[0] >= m and seg[1] >= m:
            continue

        # update counter
        if point[1] == beg_segment_const:
            counter += 1
            last_beg = point[0]
            res.add(seg)
        elif point[1] == end_segment_const:
            counter -= 1
            # discard condition
            if counter >= 1 and last_beg != point[0]:
                res.discard(seg)    
        
        # conditions when stop
        if point[1] == beg_segment_const and point[0] > m:
            break
        if counter == 0 and 0 <= point[1] <= m:
            return "No solution"

        # updates begin segment
        if point[0] < 0:
            if begin_segment_with_max_end != None:
                res.discard(begin_segment_with_max_end)
            begin_segment_with_max_end = update_begin_segment_with_max_end(begin_segment_with_max_end, seg)
            if begin_segment_with_max_end != None:
                res.add(begin_segment_with_max_end)
        
        # updates end segment
        if point[0] > m:
            if end_segment_with_min_beg != None:
                res.discard(end_segment_with_min_beg)
            end_segment_with_min_beg = update_begin_segment_with_max_end(end_segment_with_min_beg, seg)
            if end_segment_with_min_beg != None:
                res.add(end_segment_with_min_beg)
    if len(res) == 0:
        return "No solution"
    return (res, segments)

res = find_segments()
if res == "No solution":
    print(res)
else:
    print(len(res[0]))
    for seg in res[1]:
        if seg in res[0]:
            print(*seg)
```
